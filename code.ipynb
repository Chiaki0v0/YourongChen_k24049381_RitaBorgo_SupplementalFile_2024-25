```python
waveform, sample_rate = torchaudio.load("audio.mp3")
waveform = waveform.mean(dim=0, keepdim=True)
resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
waveform = resampler(waveform)
```

```python
import torch.nn as nn
import torch.optim as optim

class VRDecoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.net(x)

# embedding d=1280 output=100
model = VRDecoder(input_dim=1280, hidden_dim=512, output_dim=100)

optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()

# 训练循环
for epoch in range(1000):
    x = torch.randn(1, 1280)
    y = torch.randn(1, 100)
    pred = model(x)
    loss = criterion(pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss={loss.item():.4f}")
```

```python
'''
import torch
from transformers import PerceiverImageProcessor, PerceiverForImageClassificationConvProcessing, BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
from PIL import Image
import os, json
import numpy as np
from tqdm import tqdm

device = "cuda" if torch.cuda.is_available() else "cpu"

repo = "krasserm/perceiver-io-img-clf"
processorIMG = PerceiverImageProcessor.from_pretrained("deepmind/vision-perceiver-conv")
model = PerceiverForImageClassificationConvProcessing.from_pretrained("deepmind/vision-perceiver-conv")


tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertModel.from_pretrained("bert-base-uncased").to(device)
'''
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import os, json
import numpy as np
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity

device = "cuda" if torch.cuda.is_available() else "cpu"

model_id = "openai/clip-vit-base-patch32"
processor = CLIPProcessor.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id).to(device)
model.eval()
```

```python
def get_image_embedding(path):
    img = Image.open(path).convert("RGB")
    inputs = processor(images=img, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model.get_image_features(**inputs)
    return outputs.cpu().numpy()

def get_text_embedding(text):
    inputs = processor(text=[text], return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model.get_text_features(**inputs)
    return outputs.cpu().numpy()

def metadata_to_text(metadata):
    desc = metadata.get("description", "")
    attrs = metadata.get("attributes", [])
    attr_text = " ".join([f"{a['trait_type']}: {a['value']}" for a in attrs])
    return f"{desc} | {attr_text}".strip()
```

```python
base_dir = "/content/subset"
images_dir = os.path.join(base_dir, "img")
metadata_dir = os.path.join(base_dir, "metadata")

image_embeddings, text_embeddings = [], []

for img_file in tqdm(os.listdir(images_dir)):
    img_path = os.path.join(images_dir, img_file)
    meta_file = img_file.replace(".png", ".json")
    meta_path = os.path.join(metadata_dir, meta_file)

    if not os.path.exists(meta_path):
        continue

    with open(meta_path, "r", encoding="utf-8") as f:
        metadata = json.load(f)

    text_input = metadata_to_text(metadata)

    try:
        img_emb = get_image_embedding(img_path)
        txt_emb = get_text_embedding(text_input)
        image_embeddings.append(img_emb)
        text_embeddings.append(txt_emb)
    except Exception as e:
        print(f"Error on {img_file}: {e}")

image_embeddings = np.vstack(image_embeddings)
text_embeddings = np.vstack(text_embeddings)
```

```python
sim_matrix = cosine_similarity(text_embeddings, image_embeddings)

def recall_at_k(sim_matrix, k):
    hits = 0
    for i in range(len(sim_matrix)):
        ranked = np.argsort(sim_matrix[i])[::-1][:k]
        if i in ranked:
            hits += 1
    return hits / len(sim_matrix)

def mean_reciprocal_rank(sim_matrix):
    rr = 0
    for i in range(len(sim_matrix)):
        ranked = np.argsort(sim_matrix[i])[::-1]
        rank = np.where(ranked == i)[0][0] + 1
        rr += 1 / rank
    return rr / len(sim_matrix)

print("Recall@1:", recall_at_k(sim_matrix, 1))
print("Recall@5:", recall_at_k(sim_matrix, 5))
print("MRR:", mean_reciprocal_rank(sim_matrix))
```
